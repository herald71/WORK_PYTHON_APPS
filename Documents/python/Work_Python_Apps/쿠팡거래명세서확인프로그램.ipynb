<h1> ※ 마우스 좌표 확인 프로그램

```python
import pyautogui
import keyboard
import time

def get_mouse_position():
    positions = []
    print("마우스 커서를 원하는 위치로 이동한 후 클릭하세요.")
    print("종료하려면 'q' 키를 누르세요.")
    
    try:
        while True:
            x, y = pyautogui.position()
            position_str = f'X: {x}, Y: {y}'
            print(position_str, end='\r')
            
            if keyboard.is_pressed('q'):
                print("\n프로그램을 종료합니다.")
                break
            
            if keyboard.is_pressed('left'):
                positions.append((x, y))
                print(f"\n저장된 위치: X: {x}, Y: {y}")
                time.sleep(0.5)  # 연속 클릭 방지를 위한 대기 시간
    
    except KeyboardInterrupt:
        print("\n프로그램이 중단되었습니다.")
    
    return positions

if __name__ == "__main__":
    saved_positions = get_mouse_position()
    print("\n저장된 모든 위치:")
    for i, pos in enumerate(saved_positions, 1):
        print(f"{i}. X: {pos[0]}, Y: {pos[1]}")
```

<h1> ※ 거래명세서 확인 프로그램

```python
import cv2
import pyautogui
import numpy as np
import time

def find_and_click(image_path, threshold=0.8, y_offset=0, click_delay=1):
    """
    이미지 매칭을 통해 화면에서 해당 이미지를 찾고 클릭하는 함수
    
    Parameters:
    - image_path: 찾을 이미지의 경로
    - threshold: 매칭 임계값 (0~1 사이의 값)
    - y_offset: 클릭 위치의 y축 오프셋
    - click_delay: 클릭 후 대기 시간 (초 단위)
    
    Returns:
    - 찾고 클릭 성공 시 True, 실패 시 False
    """
    time.sleep(1)  # 화면 전환 대기 시간
    
    # 스크린샷 캡처 및 OpenCV용 변환
    screenshot = pyautogui.screenshot()
    screenshot = np.array(screenshot)
    screenshot = cv2.cvtColor(screenshot, cv2.COLOR_RGB2BGR)
    
    # 템플릿 이미지 로드
    template = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    if template is None:
        print(f"이미지를 로드할 수 없습니다: {image_path}")
        return False
    
    # 스크린샷을 그레이스케일로 변환
    screenshot_gray = cv2.cvtColor(screenshot, cv2.COLOR_BGR2GRAY)
    
    # 템플릿 매칭 실행
    result = cv2.matchTemplate(screenshot_gray, template, cv2.TM_CCOEFF_NORMED)
    _, max_val, _, max_loc = cv2.minMaxLoc(result)
    
    print(f"{image_path} 매칭 점수: {max_val:.2f}")  # 매칭 점수 출력
    
    if max_val >= threshold:
        # 클릭할 좌표 계산
        template_height, template_width = template.shape
        click_x = max_loc[0] + template_width // 2
        click_y = max_loc[1] + template_height // 2 + y_offset
        
        # 클릭 및 클릭 후 대기
        pyautogui.click(click_x, click_y)
        print(f"이미지를 클릭했습니다: 위치 ({click_x}, {click_y})")
        time.sleep(click_delay)
        return True
    else:
        print(f"이미지를 찾을 수 없습니다: {image_path}")
        return False

def main():
    """
    메인 함수, 작업을 3번 반복 수행
    """
    for i in range(4):
        print(f"===== {i+1} 번째 반복 =====")

        # 1. 첫 번째 이미지 클릭 후 2초 대기
        if not find_and_click(r"C:\Users\owner\Documents\python\images\sku.png", y_offset=-25):
            print("첫 번째 이미지를 찾지 못해 프로그램을 중단합니다.")
            break

        pyautogui.hotkey('ctrl', 'end')
        time.sleep(1)  # 추가 대기

        # 2. 두 번째 이미지 클릭 후 2초 대기
        if not find_and_click(r"C:\Users\owner\Documents\python\images\trans_check.png", threshold=0.7):
            print("두 번째 이미지를 찾지 못해 프로그램을 중단합니다.")
            break

        # 3. 특정 좌표 클릭 후 1초 대기
        pyautogui.click(1053, 191)
        time.sleep(1)

        # 4. 특정 좌표 클릭 후 "Ctrl + End" 키 입력
        pyautogui.click(1136, 191)
        time.sleep(1)
        pyautogui.hotkey('ctrl', 'end')

        # 5. 세 번째 이미지 클릭
        if not find_and_click(r"C:\Users\owner\Documents\python\images\list.png", threshold=0.7):
            print("세 번째 이미지를 찾지 못해 프로그램을 중단합니다.")
            break

        # 반복 사이에 추가적인 대기 시간 (예: 3초)
        time.sleep(2)
        
        print(f"{i+1} 번째 반복에서 작업이 성공적으로 수행되었습니다.")

    # 빨간색 볼드체로 "프로그램이 완료되었습니다." 출력
    print("\033[1m\033[31m프로그램이 완료되었습니다.\033[0m")

# 프로그램 시작
if __name__ == "__main__":
    main()
```
