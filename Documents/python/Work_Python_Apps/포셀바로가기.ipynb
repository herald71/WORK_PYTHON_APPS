# 포셀(Forsell) 바로가기 프로그램

## 프로그램 설명
포셀 웹사이트를 기본 웹 브라우저에서 자동으로 여는 간단한 바로가기 프로그램입니다.

## 주요 기능
- 포셀 웹사이트 자동 열기
- 기본 웹 브라우저 사용

## 사용 방법
1. 프로그램 실행
2. 자동으로 포셀 웹사이트가 브라우저에서 열림

## 필요 라이브러리
- webbrowser: 기본 웹 브라우저 제어 (Python 표준 라이브러리)

## 주의사항
- 기본 웹 브라우저가 설정되어 있어야 합니다

```python
import webbrowser

# URL 설정
url = "https://www.forsell.co.kr/"

# 기본 웹 브라우저에서 새 창으로 URL 열기
webbrowser.open_new(url)
```
