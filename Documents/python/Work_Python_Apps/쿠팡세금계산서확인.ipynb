# 쿠팡 세금계산서 승인 자동화 프로그램

## 프로그램 설명
쿠팡 판매자 센터에서 세금계산서 승인 작업을 자동으로 수행하는 프로그램입니다. pyautogui를 사용하여 마우스 클릭과 키보드 입력을 자동화합니다.

## 주요 기능
- 미승인 세금계산서 자동 선택
- 세금계산서 승인 버튼 클릭
- 비밀번호 자동 입력
- 확인 버튼 클릭

## 사용 방법
1. 쿠팡 판매자 센터 세금계산서 페이지를 열어둠
2. 프로그램 실행
3. 자동으로 승인 작업 수행

## 필요 라이브러리
- pyautogui: 마우스/키보드 자동화

## 주의사항
- 화면 해상도와 좌표가 일치해야 합니다
- 화면이 변경되면 좌표 수정이 필요할 수 있습니다
- 개인정보(비밀번호)가 코드에 포함되어 있으므로 보안에 주의하세요

```python
# 쿠팡 썸유 세금계산서 승인작업

import pyautogui
import time

# 1. 좌표 (1664, 662)에서 클릭하고 1초 대기합니다. (미승인계산서)
pyautogui.click(1664, 662)
time.sleep(1)

# 2. 좌표 (1664, 704)에서 클릭하고 1초 대기합니다. (첫번째 계산서)
pyautogui.click(1664, 704)
time.sleep(1)

# 3. 좌표 (435, 709)에서 클릭하고 1초 대기합니다. (승인버튼)
pyautogui.click(435, 700)
time.sleep(1)

# 4. 좌표 (1005, 694)에서 클릭한 후, "hjkim7710**"을 입력하고 1초 대기합니다.
pyautogui.click(1005, 694)
pyautogui.typewrite("hjkim7710**")
time.sleep(1)

# 5. 좌표 (961, 737)에서 클릭하고 1초 대기합니다.(선택확인)
pyautogui.click(961, 737)
time.sleep(1)

# 6. 좌표 (961, 737)에서 클릭하고 1초 대기합니다. (확인버튼)
pyautogui.click(961, 737)
time.sleep(1)


# 7. "세금계산서 승인을 완료하였습니다."라는 메시지를 출력합니다.
print("세금계산서 승인을 완료하였습니다.")
```
