```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

# 웹드라이버 매니저를 사용하여 ChromeDriver 자동 설치
service = Service(ChromeDriverManager().install())

# Chrome 드라이버를 서비스 객체를 사용하여 초기화
driver = webdriver.Chrome(service=service)

driver.implicitly_wait(3)

# 예: 페이지 열기
driver.get("https://www.livesport.com/team/manchester-united/ppjDR086/")

# 원하는 작업을 수행한 후 드라이버 종료
# driver.quit()

```

```python
from bs4 import BeautifulSoup
page = driver.page_source

bs_obj = BeautifulSoup(page, "html.parser")
bs_obj
```

```python
win = bs_obj.find_all("button",{"title":"Win"})
win
```

```python
draw = bs_obj.find_all("button",{"title":"Tie"})
draw
```

```python
lose = bs_obj.find_all("button",{"title":"Loss"})
lose
```

```python
tiewin = bs_obj.find_all("button",{"title":"Tie/Win"})
tiewin
```

```python
n_win = len(win) + len(tiewin)
```
